# Praticando os conceitos Estatísticos com o Python
      
##      Análise Exploratória de dados (AED) - Analise Descritiva
  
####      Tem por objetivo organizar, descrever e apresentar os dados, de uma determinada população,  em tabelas, gráficos e medidas de resumo.


    A  estatística descritiva, juntamente com a visualização de dados, permite descobrir informações relevantes sobre cada variável analisada, bem como relações entre elas.

    Entretanto,quando a quantidade de variáveis em análise é grande e/ou seu relacionamento complexo, normalmente não linear, não conseguimos inferir corretamente as relações existentes, seja por limitação de nossa capacidade de avaliação, seja pela limitação dos métodos estatísticos.

    Nesse momento são usadas técnicas de aprendizado de máquina para tentar descobrir as relações complexas presentes nos dados.

    Ambas as técnicas (estatística e aprendizado de máquina) atuam sobre dados, buscando extrair informações.

    A principal diferença é até onde vai nossa capacidade de interpretação e a partir de onde delegamos a interpretação para a máquina.

    Técnicas estatísticas têm a vantagem de apresentar resultados com margem de erro e nível de confiânça.
    
    Aprendizado de máquina tem a vantagem de conseguir tratar relacionamentos complexos.

### Cálculo da Média Aritmética de uma variável x:




```python
# Média Aritmética
# Soma de valores dividido por n

# idade dos colaboradores da empresa x

X=[25,32,29,45]


print('soma',sum(X)) 

print('qte de observações',len(X)) 
 
Média_x = sum(X)/len(X)

print('média', Média_x)

```

```python
# idade dos colaboradores da empresa y

y=[60,25,32,29,45]

Média_y = sum(y)/len(y)

Média_y
```

```python
# utilizando a biblioteca numpy

np.mean(X)
```

```python
np.mean(y)
```

 ####  A mediana é uma medida que, como média, também procura caracterizar o centro da distribuição da frequência.
 
     Ela é calculada com base na ordem dos valores que formam o conjunto dos dados.
     Os modelos são diferentes para situações onde existem um número ímpar ou par de observações.
 
 
 
 
      Mediana: definida como o valor central das observações ordenadas quando n for ímpar. 
      Para oscasos em que n for par, a mediana é definida como valor médio entre as duas observações centrais ordenadas
      
      
      
      

```python
# Calculando a Mediana da variável x :

# lembrando ordenação x = np.sort(x)  # coloca os valores do vetor em ordem crescente.

print('Mediana',np.median(X))

```

```python
# Calculando a Mediana da variável y :

print('Mediana',np.median(y))
```

```python
# Medidas de Dispersão
# Produção diaria de dois empregados

EmpregA = [70,71,69,72,70,71,72]
EmpregB = [60,80,70,62,75,71,83] 

# Média de produção
    print ('média da produção', '\n Empregado A:', np.mean(EmpregA),'\n Empregado B:', np.mean(EmpregB))

# Mediana de produção
print ('\n mediana da produção', '\n Empregado A:', np.median(EmpregA),'\n Empregado B:', np.median(EmpregB))
```

```python
# utilizando a biblioteca matplotlib.pyplot

#plt.plot(EmpregA, color='g')
plt.plot(EmpregA, 'g:')
plt.plot(EmpregB, 'b:')
plt.ylabel('Produção')
plt.xlabel('Tempo')

plt.xticks([])
plt.legend(['Empregado A','Empregado B'], loc='best')
sb.despine() #remover o quadro 
```

```python
# Amplitude: Avaliar a dispersão do conjunto de dados

# Amplitude= Max(dados)- Min(dados)

Ampl_A= max(EmpregA)- min(EmpregA)

Ampl_B= max(EmpregB)- min(EmpregB)


# Amplitude de produção
print ('Amplitude de Produção de cada empregado', '\n Empregado A:', Ampl_A,'\n Empregado B:', Ampl_B)
```

#### Medidas de Dispersão:

    As medidas de dispersão são medidas que indicam o quão espalhados os dados estão, ou como eles variam. 
    
    São elas a amplitude, a variância, o desvio padrão e o desvio absoluto. 
    
    Variância: uma medida que expressa quanto os dados de um conjunto estão afastados de seu valor esperado.
    
    
    
    Calculamos a variância no Pandas através da função var():
    

     

```python
# calculando a distancia entre cada ponto é a media 

dist_A = EmpregA - np.mean(EmpregA)

dist_A
```

```python
print('soma da distancia entre os pontos e a média Empregado A',round(sum(dist_A),10)) 
```

```python
print('soma da distancia absoluta  entre os pontos e a média Empregado A',sum(abs(dist_A))) 
```

```python
print('média da distancia entre os pontos e a média Empregado A', sum(abs(dist_A))/len(EmpregA))
```

```python
# calculando a distancia entre cada ponto é a media 

dist_B = EmpregB - np.mean(EmpregB)

dist_B
```

```python
print('soma da distancia entre os pontos e a média Empregado B',sum(dist_B)) 
```

```python
print('média da distancia absolutaentre os pontos e a média Empregado B', sum(abs(dist_B ))/len(EmpregB))
```

```python

# Desvio Padrão:
# uma medida de dispersão, que indica quanto os dados estão afastados da média. 
# Um valor de desvio padrão alto indica que os valores estão mais espalhados, mais longe da média, e um desvio padrão
# baixo indica que os valores estão mais próximos da média. 
# No Pandas fazemos o cálculo através da função std():

dpA=np.std(EmpregA)

dpB=np.std(EmpregB)

 
# print ('desvio padrão dos empregados','\n Empregado A:', {dpA,:.3f} ,'\n Empregado B:', {dpB, :.3f})
# pd.set_option('float_format','{:.3f}'.format)

print ('desvio padrão dos empregados','\n Empregado A:', dpA , '\n Empregado B:',dpB)
```

#### Coeficiente de Variação 

    Usado quando queremos comparar duas variáveis quantitativas quanto ao seu grau de dispersão.
    
    É o quociente entre o desvio-padrão e a média.
    
    



    Vantagem: caracterizar a dispersão dos dados em termos relativos a seu valor médio.


```python
# Coeficiente de Variação

CV_A = np.std(EmpregA)/np.mean(EmpregA)*100
CV_B = np.std(EmpregB)/np.mean(EmpregB)*100
 

# Coeficiente de Variação
print ('Coeficiente de Variação com relaçao a Produção de cada empregado', '\n Empregado A:', CV_A ,'\n Empregado B:',CV_B )
```

    Variável aleatória:
    
    É uma variável cujo valor é um resultado numérico, associado ao resultado de uma experiência aleatória.
    
    É a expressão matemática de eventos sujeitos à variação natural (do acaso), seus resultados são expressos 
    
    por conjuntos de um espaço amostral, e possuirá sempre uma distribuição de probabilidade.
